# DATA VISUALISATION

```python
#Terrorist Activity in each year
plt.figure(figsize = (15,5))
ax = sns.countplot(x='Year',data=df)
plt.xticks(rotation=90)
# Adding annotations to the chart
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='baseline', fontsize=12, color='black', xytext=(0, 5),
                textcoords='offset points',rotation = 90)
    
plt.title('Number of terrorist activites each year')
plt.show
```

Year with lowest Terrorist Activities : 1971 and the year with highest Terrorist Activities : 2014.

```python
#Most Affected Region Due to Terrorist Attack
region_counts = df['Region'].value_counts()

# Step 1: Create the pie chart
plt.figure(figsize=(10, 8))  
ax = plt.gca()
pie = plt.pie(region_counts, labels=None, autopct='%1.1f%%', startangle=140)
plt.title('Terrorist Incidents by Region')

# Step 2: Add labels with a legend to the left side
ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Add a legend to the left side
plt.legend(region_counts.index, loc='center left', bbox_to_anchor=(0.9, 0.5))

# Center the pie chart
plt.subplots_adjust(left=0.0, right=0.7)  # Adjust the right value as needed
plt.show()
```

Middle East and North Africa are the most affected Region with the highest incidence of terrorist attacks.

```python
#Terrorist Acitivity increasing over the time in each region
region_year_counts = df.groupby(['Region', 'Year']).size().unstack(fill_value=0)
plt.figure(figsize=(12, 8))

for region in region_year_counts.index:
    plt.plot(region_year_counts.columns, region_year_counts.loc[region], label=region)

plt.title('Terrorist Activity Over Time in Each Region')
plt.xlabel('Year')
plt.ylabel('Number of Incidents')
plt.legend(loc='upper left')
plt.grid(True)

plt.show()
```

The above plot clearly illustrates a consistent and concerning trend of escalating terrorist activities in the Middle East and North Africa, as well as the South Asia region. These regions consistently experience the highest levels of terrorist incidents, and this upward trajectory persists over time.

```python
#Terroritst Attack Per country
fig,ax = plt.subplots(figsize=(10,5))
ax = sns.barplot(x=df.Country.value_counts()[:20].values,y = df.Country.value_counts()[:20].index)
ax.set_title('Terrorist Attacks per Country')
```

Count of terrorist attacks is more in Iraq, Pakisthan and Afghanisthan

```python
#Most Common Attack Method
plt.figure(figsize=(16,3))
sns.countplot(x = df['Attacktype'], order = df['Attacktype'].value_counts().index)
plt.xticks(rotation=60)
plt.xlabel('Method')
plt.title('Method of Attack')
plt.show()
```

The prevailing pattern in terrorist attacks points to a significant reliance on the methods of bombing and assassination.

```python
#Killed Vs Wounded in Each Attacktype
attack_data = df.groupby('Attacktype')[['killed', 'Wounded']].sum()
attack_data.plot(kind='bar', stacked=True,figsize = (10,6))
plt.xlabel('Attack Type')
plt.ylabel('Count')
plt.title('Fatalities vs. Injuries by Attack Type')
plt.show()
```

```python
#Lethality Ratio 
attack_data['killed'] / attack_data['Wounded']
```

**Armed assault** and **Assassination** were identified as the most lethal attack types, exhibiting the highest lethality ratios among all recorded incidents.

```python
#Most Active Terrorist organisation
df.Group.value_counts()[:20]
```

```python
plt.subplots(figsize=(11,5))
sns.barplot(y=df.Group.value_counts()[1:20].index,x=df.Group.value_counts()[1:20].values)
plt.title('Most Active Terrorist Organizations')
plt.show()
```

Taliban and ISL are the most active groups which have attacked more than 7000 times

```python
#Terrorist Group Activity over the years
group_activity = df.groupby(['Year', 'Group']).size().reset_index(name='incident_count')
most_active_groups = group_activity.groupby('Year')['Group', 'incident_count'].max().reset_index()


plt.figure(figsize=(12, 6))
for group in most_active_groups['Group'].unique():
    group_data = most_active_groups[most_active_groups['Group'] == group]
    plt.plot(group_data['Year'], group_data['incident_count'], label=group)
    
plt.title('Most Active Terrorist Groups Over the Years')
plt.xlabel('Year')
plt.ylabel('Incident Count')
plt.legend(loc='best')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)
plt.show()
```

We Can see by the above line graph that White Extremists is the most active group in the early 1970s and after that
Zimbawe African Peoples unions become the most acitve terrorist group 

```python
df['Casualties'] = df['killed'] + df['Wounded']
```

```python
df.Casualties.isnull().sum()
```

```python
plt.subplots(figsize = (15,6))
cas = df.groupby('Year')['Casualties'].sum().to_frame().reset_index()
cas.columns = ['Year','Casualties']
sns.barplot(x=cas['Year'],y=cas['Casualties'])
plt.xticks(rotation = 90)
plt.show()
```

```python
cas_max_year = cas[cas['Casualties'] == cas['Casualties'].max()]['Year'].iloc[0]
cas_min_year = cas[cas['Casualties'] == cas['Casualties'].min()]['Year'].iloc[0]
print(f"Year with Maximum Casualties: {cas_max_year} and maximum Casualties is: {cas['Casualties'].max()}")
print(f"Year with Minimum Casualties: {cas_min_year} and maximum Casualties is: {cas['Casualties'].min()}")
```

```python
#Casualities Cause by Each Terrorist Group
group_casualties = df.groupby('Group')['Casualties'].sum().reset_index()
top_10_dangerous_group = group_casualties.sort_values(by = 'Casualties',ascending =False).head(10)
```

```python
plt.subplots(figsize=(11,5))
sns.barplot(y=top_10_dangerous_group['Group'][1:],x=top_10_dangerous_group['Casualties'][1:])
plt.title('Most Harmful Terrorist Organizations')
plt.show()
```

The analysis reveals that ISIL and the Taliban are the organizations responsible for the maximum casualties.

```python
#Most Affected Countries due to Terrorist attack
plt.subplots(figsize = (15,8))
country_killed = df.groupby('Country').killed.sum().to_frame().reset_index().sort_values('killed',ascending = False)[:15]
ax = sns.barplot(x = country_killed.Country, y= country_killed.killed, palette = "GnBu_r",edgecolor = 
                 sns.color_palette('dark',10))
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='baseline', fontsize=12, color='black', xytext=(0, 5),
                textcoords='offset points',rotation = 45)

plt.show()
```

Terrorist acts in the Middle East and northern Africa have been seen to have fatal consequences. The Middle East and North Africa are seen to be the places of serious terrorist attacks. In addition, even though there is a perception that Muslims are supporters of terrorism, Muslims are the people who are most damaged by terrorist attacks. If you look at the graphics, it appears that Iraq, Afghanistan and Pakistan are the most damaged countries. All of these countries are Muslim countries.
